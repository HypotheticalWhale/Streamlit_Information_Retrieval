## Extracting quarters

```python
frame['Datetime'] = frame['Datetime'].map(lambda Datetime: Datetime.split()[0])

frame
```

```python
frame['Datetime'] = frame['Datetime'].apply(pd.to_datetime)

frame
```

```python
def check_quarter(target_date):
  # 2021
  first_start_21 = datetime.datetime.strptime("2021-01-01", "%Y-%m-%d")
  first_end_21 = datetime.datetime.strptime("2021-03-31", "%Y-%m-%d")

  second_start_21 = datetime.datetime.strptime("2021-04-01", "%Y-%m-%d")
  second_end_21 = datetime.datetime.strptime("2021-06-30", "%Y-%m-%d")

  third_start_21 = datetime.datetime.strptime("2021-07-01", "%Y-%m-%d")
  third_end_21 = datetime.datetime.strptime("2021-09-30", "%Y-%m-%d")

  fourth_start_21 = datetime.datetime.strptime("2021-10-01", "%Y-%m-%d")
  fourth_end_21 = datetime.datetime.strptime("2021-12-31", "%Y-%m-%d")

  # 2022
  first_start_22 = datetime.datetime.strptime("2022-01-01", "%Y-%m-%d")
  first_end_22 = datetime.datetime.strptime("2022-03-31", "%Y-%m-%d")

  second_start_22 = datetime.datetime.strptime("2022-04-01", "%Y-%m-%d")
  second_end_22 = datetime.datetime.strptime("2022-06-30", "%Y-%m-%d")

  third_start_22 = datetime.datetime.strptime("2022-07-01", "%Y-%m-%d")
  third_end_22 = datetime.datetime.strptime("2022-09-30", "%Y-%m-%d")

  fourth_start_22 = datetime.datetime.strptime("2022-10-01", "%Y-%m-%d")
  fourth_end_22 = datetime.datetime.strptime("2022-12-31", "%Y-%m-%d")

  # 2023
  first_start_23 = datetime.datetime.strptime("2023-01-01", "%Y-%m-%d")
  first_end_23 = datetime.datetime.strptime("2023-03-31", "%Y-%m-%d")

  # 2021
  if first_start_21 <= target_date <= first_end_21:
    return 11
  elif second_start_21 <= target_date <= second_end_21:
    return 12
  elif third_start_21 <= target_date <= third_end_21:
    return 13
  elif fourth_start_21 <= target_date <= fourth_end_21:
    return 14
  
  # 2022
  elif first_start_22 <= target_date <= first_end_22:
    return 21
  elif second_start_22 <= target_date <= second_end_22:
    return 22
  elif third_start_22 <= target_date <= third_end_22:
    return 23
  elif fourth_start_22 <= target_date <= fourth_end_22:
    return 24
  
  # 2023
  elif first_start_23 <= target_date <= first_end_23:
    return 31
```

```python
frame['Quarter'] = frame['Datetime'].map(lambda Datetime: check_quarter(Datetime))

frame
```
