## Further preprocessing

```python
def clean_urls(review):
    review = review.split()
    review = ' '.join([word for word in review if not re.match('^http', word)])
    return review
```

```python
def clean_text(text):
    text = str(text)
    text = re.sub(r'(\w)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-zA-Z0-9 ]+', ' ', text)
    text = re.sub(r'http\S+', ' ', text)
    text = re.sub(r'https?:\/\/.*[\r\n]*', '', text)
    text = re.sub(r'^RT[\s]+', '', text)
    text = re.sub(r'pic.twitter\S+', ' ', text)
    text = re.sub(r'#', '', text)
    text = re.sub(r'@\w+', '', text)
    text = text.lower()

    return text
```

```python
stop_words = stopwords.words('english') 
stop_words.remove('not') 
lemmatizer = WordNetLemmatizer()
```

```python
def data_preprocessing(review):
    
  # data cleaning
    review = re.sub(re.compile('<.*?>'), '', review) #removing html tags
    review =  re.sub('[^A-Za-z0-9]+', ' ', review) #taking only words
  
  # lowercase
    review = review.lower()
  
  # tokenization
    tokens = nltk.word_tokenize(review) # converts review to tokens
  
  # stop_words removal
    review = [word for word in tokens if word not in stop_words] #removing stop words
  
  # lemmatization
    review = [lemmatizer.lemmatize(word) for word in review]
  
  # join words in preprocessed review
    review = ' '.join(review)
    return review
```

```python
clean_df['Clean Text'] = clean_df['Text'].apply(clean_urls).apply(clean_text)
```

```python
clean_df['Clean Text'] = clean_df['Clean Text'].apply(lambda review: data_preprocessing(review))

clean_df
```

## Vader classifier

```python
#!python -m spacy download en_core_web_sm
```

```python
import en_core_web_sm
import random
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
```

```python
def vader_analysis(sentence):
  classifier = SentimentIntensityAnalyzer()
  result = classifier.polarity_scores(sentence)
  score = result['compound']

  if score >= 0.05: # pos
    result = 'pos'
  elif score > -0.05 and score < 0.05: # neu
    result = 'neu'
  else: # neg
    result = 'neg'
  
  return result
```

```python
clean_df['vader'] = clean_df['Clean Text'].apply(vader_analysis)

clean_df
```

```python
vader_counts = clean_df['vader'].value_counts()

vader_counts
```

## Textblob classifier

```python
def textblob_analysis(sentence):
  score = TextBlob(sentence).sentiment.polarity

  if score < 0:
      result = 'neg'
  elif score == 0:
    result = 'neu'
  else:
    result = 'pos'

  return result
```

```python
clean_df['textblob'] = clean_df['Clean Text'].apply(textblob_analysis)

clean_df
```

```python
textblob_counts = clean_df['textblob'].value_counts()

textblob_counts
```
