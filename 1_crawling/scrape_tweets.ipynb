Crawling of tweets for 10 nfts using snscrape

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

## NFTs

* Mutant Ape Yacht Club (MAYC)
* Azuki
* Bored Ape Yacht Club (BAYC)
* Meebits
* CryptoPunks
* CloneX
* The Potatoz
* Pixelmon
* MekaVerse
* Phanta Bear

```python
def to_csv(t_df, t_name): 
    t_df.to_csv(t_name, sep=',', index=False)
```

### Pixelmon

```python
pixel_query = ['pixelmon nft since:2023-01-01 until:2023-03-31 exclude:retweets exclude:replies',
              'pixelmon nft since:2022-10-01 until:2022-12-31 exclude:retweets exclude:replies',
              'pixelmon nft since:2022-07-01 until:2022-09-30 exclude:retweets exclude:replies',
              'pixelmon nft since:2022-04-01 until:2022-06-30 exclude:retweets exclude:replies',
              'pixelmon nft since:2022-01-01 until:2022-03-31 exclude:retweets exclude:replies'
              ]
          
t_max = 500
t_list = []
for query in pixel_query:
  for i,tweet in enumerate(sntwitter.TwitterSearchScraper(query).get_items()):
          if i>t_max:
              break
          t_list.append([tweet.date, tweet.id, tweet.rawContent, tweet.user.username, tweet.replyCount, tweet.retweetCount ,tweet.likeCount])

pixel_df = pd.DataFrame(t_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username', 'Replies', 'Retweets', 'Likes'])
to_csv(pixel_df, 'pixel_tweets.csv')

pixel_df
```
